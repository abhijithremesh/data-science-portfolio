```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# The dataset contains Covid-19 patients in India from Jan till 5th April 2020

```python
cases = pd.read_csv('Covid cases in India.csv',index_col=0)
coord = pd.read_csv('Indian Coordinates.csv',index_col=0)
daily = pd.read_excel('per_day_cases.xlsx',index_col=0)
```

```python
cases.head()
```

#  Confirmed cases vs States of India

```python
plt.figure(figsize=(12,15))
sns.barplot(x='Total Confirmed cases',y='Name of State / UT',data=cases)
plt.savefig('images/Confirmed cases vs States of India.png')
```

#  Cured cases vs States of India

```python
plt.figure(figsize=(12,15))
sns.barplot(x='Cured/Discharged/Migrated',y='Name of State / UT',data=cases)
plt.savefig('images/Cured cases vs States of India.png')
```

#  Death cases vs States of India

```python
plt.figure(figsize=(12,15))
sns.barplot(x='Deaths',y='Name of State / UT',data=cases)
plt.savefig('images/Death cases vs States of India.png')
```

# (Total Cases , New Cases) vs Time

```python
Date = daily['Date']
TotalCases = daily['Total Cases']
NewCases = daily['New Cases']
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.title('Total cases vs time')
plt.tight_layout()
plt.xticks(rotation=90)
plt.plot(Date,TotalCases,'b')
plt.subplot(1,2,2)
plt.title('New cases vs time')
plt.tight_layout()
plt.plot(Date,NewCases,'g')
plt.xticks(rotation=90)


```
